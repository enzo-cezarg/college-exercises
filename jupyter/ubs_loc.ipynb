```python
# Inicializa um dicionário vazio
unidades = {}

# Abre o arquivo csv que contém informações sobre todas as UBS do país
with open('../csv/ubs.csv') as csv:
  # Pula a primeira linha
  print(csv.readline())

  # Para cada linha do arquivo:
  for linha in csv:
    # Divide os valores da linha e corta o último caractere (\n)
    aux = linha[:-1].split(';')
    lat = aux[6]
    lon = aux[7]
    uf = aux[1]

    # Percorre as latitudes e substitui as vírgulas por pontos
    for pos in range(0, len(lat)):
      if lat[pos] == ',':
        # Concatenação de duas strings fatiadas com o caractere desejado
        lat = lat[:pos] + '.' + lat[pos+1:]

    # Percorre as longitudes e substitui as vírgulas por pontos
    for pos in range(0, len(lon)):
      if lon[pos] == ',':
        # Concatenação de duas strings fatiadas com o caractere desejado
        lon = lon[:pos] + '.' + lon[pos+1:]

    if lat != '' and lon != '' and uf == '43': # Filtrando unidades no RS
      nome = aux[3][1:-1]
      logr = aux[4][1:-1]
      bairro = aux[5][1:-1]
      # Conversão das strings para floats
      lat = float(lat)
      lon = float(lon)
      unidades[nome] = (logr, bairro, lat, lon)

encontrou = False
# Consulta por chave
ubs = input('Digite o nome da UBS (sem acentuação): ').upper()
print('===='*10)

if ubs in unidades:
  print(unidades[ubs])
  encontrou = True
else:
  # Consulta parcial  
  for nome in unidades.keys():
    if ubs in nome:
      print(f'Parcial: {nome}, {unidades[nome]}')
      encontrou = True

if encontrou == False:
  print('Não encontrado!')
  print('===='*10)

# Fecha o arquivo    
csv.close()
```
